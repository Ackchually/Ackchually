```python
import re
import socket

def is_public_ip(ip_address):
    private_ip_regex = re.compile(r'^10\.|^172\.(1[6-9]|2\d|3[01])\.|^192\.168\.')
    return not private_ip_regex.match(ip_address)

def get_public_ips(ip_addresses):
    public_ips = []
    for ip_address in ip_addresses:
        if is_public_ip(ip_address):
            public_ips.append(ip_address)
    return public_ips

# Example usage:
ip_addresses = input("Please paste the IP addresses separated by commas: ").split(",")
public_ips = get_public_ips(ip_addresses)
```

```python
for ip_address in public_ips:
    print(ip_address)
```
